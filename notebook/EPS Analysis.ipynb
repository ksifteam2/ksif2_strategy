```python
sector_code = pd.read_csv("../data/firm_sector_code.csv")
```

```python
eps = eps.T
```

```python
sector_code = sector_code.rename(columns={'Firm': 'Symbol'})
sector_code = sector_code.set_index('Symbol')
sector_code.index.name = 'Symbol'
eps.index.name = 'Symbol'
```

```python
eps = eps.join(sector_code)
```

```python
for grp in eps.groupby('산업명-대분류'):
    print(grp[0])
    fig = plt.figure()
    ax = fig.add_subplot(111)
    ax.hist(grp[1].T.iloc[1:-5].mean().dropna().values)
#     print(grp[1].T.iloc[1:-5].quantile([0.25, 0.5, 0.75]))
    print(grp[1].T.iloc[1:-5].mode())
```
